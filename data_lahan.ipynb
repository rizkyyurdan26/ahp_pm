merge all dataset

```python
import pandas as pd
import os

# Folder tempat file Excel berada
folder_path = './clean'  # Ganti dengan path folder Anda
output_file = 'merged_data.xlsx'  # Nama file yang ingin ditimpa

# Membuat folder jika belum ada
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Inisialisasi list untuk menyimpan semua DataFrame
all_data = []

# Loop melalui semua file Excel di folder
for file in os.listdir(folder_path):
    if file.endswith('.xlsx') and file != output_file:  # Hindari membaca file output
        file_path = os.path.join(folder_path, file)
        # Membaca file Excel
        df = pd.read_excel(file_path)
        all_data.append(df)

# Menggabungkan semua DataFrame menjadi satu
merged_data = pd.concat(all_data, ignore_index=True)

# Menyimpan DataFrame gabungan ke file Excel baru (menimpa jika sudah ada)
output_path = os.path.join(folder_path, output_file)
merged_data.to_excel(output_path, index=False)

print(f"File '{output_file}' telah berhasil ditimpa dengan data gabungan.")
```

### Tambah Data

```python
import pandas as pd

df_kurang = pd.read_excel('./dataset/dataset/data kurang.xlsx')


df_kurang.head(10)
```

```python
df_kurang.drop(columns=['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7'], axis= 1, inplace=True)
df_kurang.head()
```

```python
df_kurang.rename(columns={
    'Nama_Lahan' : 'nama_lahan',
    'Kemiringan' : 'kemiringan',
    'Drainase (lancar, terhambat, lambat' : 'drainase',
    'pH': 'ph',
    'Tekstur tanah': 'tekstur_tanah',
}, inplace=True)

df_kurang.head(10)
```

```python
#mengubah data %
def calculate(value):
    value = str(value)
    if '-' in value:
        start, end = value.replace('%','').split('-')
    elif ' - ' in value:
        start, end = value.split(' - ')
    else:
        num_value = float(value)

        # Jika nilai kurang dari 1, kalikan 100
        if num_value < 1:
            return num_value * 100
        return num_value

    # Mengembalikan rata-rata dari start dan end
    return (float(start.strip()) + float(end.strip())) / 2

# Memeriksa apakah kolom 'kemiringan' ada, lalu mengaplikasikan fungsi calculate
if 'kemiringan' in df_kurang.columns:
    df_kurang['kemiringan'] = df_kurang['kemiringan'].apply(calculate)

# Memeriksa apakah kolom 'ph' ada, lalu mengaplikasikan fungsi calculate
if 'ph' in df_kurang.columns:
    df_kurang['ph'] = df_kurang['ph'].apply(calculate)

df_kurang.head(10)
```

```python
df_kurang['drainase'] = df_kurang['drainase'].str.lower()

# Mengubah kolom 'tekstur_tanah' dengan menghapus kata 'Tekstur' dan mengubahnya menjadi huruf kecil
df_kurang['tekstur_tanah'] = df_kurang['tekstur_tanah'].str.replace('Tekstur', '', regex=False).str.lower()


df_kurang.head(10)
```

```python
df_kurang.to_excel('./clean/data_kurang.xlsx', index=False)
```
