```python
import nbformat
import pandas as pd

# Fungsi untuk memuat variabel dari notebook
def load_variable_from_notebook(notebook_path, variable_name):
    # Membuka dan membaca notebook
    with open(notebook_path, 'r', encoding='utf-8') as f:
        notebook = nbformat.read(f, as_version=4)
    
    # Membuat namespace baru untuk mengeksekusi kode
    namespace = {}
    
    # Mengeksekusi setiap sel kode dalam notebook
    for cell in notebook.cells:
        if cell.cell_type == 'code':
            exec(cell.source, namespace)
    
    # Mengambil variabel yang diminta dari namespace
    variable = namespace.get(variable_name)
    if variable is None:
        raise ValueError(f"Variable '{variable_name}' not found in {notebook_path}")
    return variable
```

# Ranking Varietas MM

```python


# Mengimpor variabel dari masing-masing notebook
total_skor_tanah_MM = load_variable_from_notebook('pm_tanah.ipynb', 'total_skor_tanah_MM')
total_skor_topo_MM = load_variable_from_notebook('pm_topografi.ipynb', 'total_skor_topo_MM')
skor_total_iklim_MM = load_variable_from_notebook('pm_iklim.ipynb', 'skor_total_iklim_MM')  # Ganti nama variabel yang dimuat

# Membuat DataFrame dari setiap variabel yang telah diimpor
import pandas as pd

# Menggabungkan DataFrame berdasarkan 'nama_lahan'
df_tanah = total_skor_tanah_MM[['nama_lahan', 'total_skor_tanah_MM']]
df_topo = total_skor_topo_MM[['nama_lahan', 'total_skor_topo_MM']]
df_iklim = skor_total_iklim_MM[['nama_lahan', 'total_skor_iklim_MM']]  # Pastikan nama variabel sesuai

df_combined = pd.merge(df_iklim, df_topo, on='nama_lahan')
df_combined = pd.merge(df_combined, df_tanah, on='nama_lahan')

# Menambahkan kolom baru untuk total skor keseluruhan
df_combined['total_skor_keseluruhan'] = (
    df_combined['total_skor_iklim_MM'] + 
    df_combined['total_skor_topo_MM'] + 
    df_combined['total_skor_tanah_MM']
)

# Mengurutkan DataFrame berdasarkan 'total_skor_keseluruhan' secara descending
df_combined_sorted = df_combined.sort_values(by='total_skor_keseluruhan', ascending=False)

# Menambahkan kolom ranking
df_combined_sorted['ranking'] = range(1, len(df_combined_sorted) + 1)

# Menyusun ulang kolom sehingga 'ranking' berada setelah 'total_skor_keseluruhan'
cols = df_combined_sorted.columns.tolist()
col_pos = cols.index('total_skor_keseluruhan') + 1  # Dapatkan posisi setelah 'total_skor_keseluruhan'
cols.insert(col_pos, cols.pop(cols.index('ranking')))  # Pindahkan 'ranking' ke posisi yang diinginkan

# Menerapkan urutan kolom baru
df_combined_sorted = df_combined_sorted[cols]

# Menampilkan DataFrame yang telah disusun
df_combined_sorted
```
