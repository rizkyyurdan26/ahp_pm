```python
import pandas as pd
```

```python
df = pd.read_excel('./dataset/dataset/data_jagung.xlsx')
df
```

```python
df.columns
```

```python
df.drop(columns=['suhu_avg', 'curah_hujan_avg', 'kelembapan_avg','Bentuk lahan(bukit, lereng, gunung, dataran)', 'bahan organik'], axis = 1, inplace=True)
df
```

```python
df.columns
```

```python
df.rename(columns={
    'Jenis Jagung' : 'jenis_varietas',
    'kelembapan_max': 'kelembapan_udara_max',
    'kelembapan_min': 'kelembapan_udara_min',
    'durasi_matahari_jam' : 'penyinaran',
    'kelembapan_tanah_avg' : 'kelembapan_tanah',
    'Jenis Tanah' : 'jenis_tanah',
    'Kemiringan' : 'kemiringan',
    "Drainase (lancar, terhambat, lambat": 'drainase',
    'pH':'ph',
    'Tekstur tanah':'tekstur_tanah',
    
}, inplace=True)

df
```

Drop symbol


```python
df['curah_hujan_max'] = df['curah_hujan_max'].str.replace(' MM/Tahun', '', regex=False)
df['curah_hujan_min'] = df['curah_hujan_min'].str.replace(' MM/Tahun', '', regex=False)
df['penyinaran'] = df['penyinaran'].str.replace(' Jam/Hari', '', regex=False)
df['ketinggian'] = df['ketinggian'].str.replace(' MDPL', '', regex=False)
df['kelembapan_tanah'] = df['kelembapan_tanah'].str.replace('%', '', regex=False)
df['kemiringan'] = df['kemiringan'].str.replace('0- ', '', regex=False)
df
```

Split data


```python
df[['kelembapan_tanah_min', 'kelembapan_tanah_max']] = df['kelembapan_tanah'].str.split('-', expand=True)

df[['ketinggian_min', 'ketinggian_max']] = df['ketinggian'].str.split('-', expand=True)

df[['penyinaran_min','penyinaran_max']] = df['penyinaran'].str.split('-', expand=True)

df[['ph_min','ph_max']] = df['ph'].str.split('-', expand=True)

df= df.drop(columns=['kelembapan_tanah', 'ketinggian', 'penyinaran', 'ph'], axis = 1, )

df.head()
```

```python
df['kelembapan_udara_max'] = df['kelembapan_udara_max'] * 100
df['kelembapan_udara_min'] = df['kelembapan_udara_min'] * 100

df
```

check dtypes

```python
df.dtypes
```

Convert dtypes

```python
data_convert = ['suhu_max', 'suhu_min', 'curah_hujan_max', 'curah_hujan_min', 'kelembapan_udara_max', 'kelembapan_udara_min', 'kemiringan', 'kelembapan_tanah_max', 'kelembapan_tanah_min','ketinggian_max', 'ketinggian_min', 'penyinaran_max', 'penyinaran_min', 'ph_max', 'ph_min', ]

df[data_convert] = df[data_convert].astype(float)
df.head()
```

```python
df.dtypes
```

```python
df.to_excel('./dataset/data_varietas.xlsx', index=False)
```

```python
import pandas as pd

df = pd.read_excel('./dataset/data_varietas.xlsx')
df.head()
```
